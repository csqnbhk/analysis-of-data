# 关于笑话网的数据分析


## 1.需求分析
- 为http://www.jokeji.cn/这个网址的建设者提供有效的网站建议。
- 获取该网站的所有笑话类型,每个笑话的访问量.分析游览者喜欢看什么类型的笑话及用户的一些喜好。为该网站的长久建设提供有效的建议。

## 2.数据获取
-  获取 http://www.jokeji.cn/Keyword.htm这个入口下的所有笑话类型和该类型的笑话数量及每种类型笑话的入口URL,写入jokes_data_analysis数据库的joke_type表中。
- 从每种笑话类型的的入口URL,获取每种笑话类型的页数和每个笑话的标题,URL和访问量。把获取的信息写入jokes_data_analysis数据库的joke_more_info表中。



# 3.数据分析

### 1. 先连接数据库

### - 查看一下joke_type表存的什么数据


```python
joke_style = pd.read_sql("select * from joke_type order by page desc", con=engine)
print('该表的数据条目为:%d'% len(joke_style))
print(joke_style)
```

### - 查看一下joke_more_info表存的什么数据

```python
joke_more_info= pd.read_sql("select * from joke_more_info order by visit desc", con=engine)
print('该表的数据条目为:%d'% len(joke_more_info))
print(joke_more_info)
```

# 1.分析joke_type表的数据

```python
print('类型→数量')
for (style,number) in zip(joke_style['style'],joke_style['number']):
    print('%s→%d' %(style[2:],number))
```

## 笑话类型对应的笑话数量直方图如下:


```python
## 设置中文显示
## 设置字体为SimHei显示中文
plt.rcParams["font.sans-serif"] = "Microsoft YaHei"
## 设置正常显示符号
plt.rcParams["axes.unicode_minus"] = False
## 从文件导入数据
data = joke_style;
## 提取其中的columns数组，视为数据的标签
name = []
for style in joke_style['style']:
    name.append(style[2:])
## 提取其中的values数组，数据的存在位置
values = data["number"]
#print(name)
#print(values)
# 刻度标签
labels = name
## 设置画布
plt.figure(figsize=(20,10))
## 绘制直方图
plt.bar(range(len(data)), values, width=0.8, color="r")
## 添加横轴标签
plt.xlabel("类型")
## 添加y轴名称
plt.ylabel("数量")
## 设置x轴刻度
plt.xticks(range(len(data)), labels)
## 添加图表标题
plt.title("笑话类型及其数量直方图")
plt.xticks(range(len(labels)),name[:], rotation=45)
## 保存图形到文件
plt.savefig("笑话类型及其数量直方图.png")
## 显示图形
plt.show()
```

## 笑话类型对应的笑话数量饼图如下:

```python
plt.figure(figsize=(20,20))
plt.pie(values, labels=name[:], autopct="%1.2f%%")
## 保存图形到文件
plt.savefig("笑话类型及其数量饼图.png")
plt.show()
```

# 2.分析joke_more_info表数据

### - 先清除一些没用的数据,获取对现在分析有用的数据

```python

list_type = []
data = []
for i in joke_more_info['type']:
    list_type.append(i[2:])
for (j_title,j_visit,j_type)  in zip(joke_more_info['title'],joke_more_info['visit'],list_type):
    data.append([j_title,j_visit,j_type])
print('笑话题目→访问量→类型')    
for (j_title,j_visit,j_type) in data:
    print('{}→{}→{}'.format(j_title,j_visit,j_type))
```

### - 绘制直方图,该图表示每种类型笑话的平均访问量。

```python
#data=np.array(data)
#print(joke_more_info)
type_meanvisit = pd.read_sql("select type,avg(visit) from joke_more_info group by type order by avg(visit) desc", con=engine)
#print('类型→平均访问量')
for (j_type,meanvisit) in zip(type_meanvisit['type'],type_meanvisit['avg(visit)']):
    print('%s→%d' %(j_type[2:],meanvisit))
```

## - 笑话类型对应笑话的平均访问量直方图

```python
## 设置中文显示
## 设置字体为SimHei显示中文
plt.rcParams["font.sans-serif"] = "Microsoft YaHei"
## 设置正常显示符号
plt.rcParams["axes.unicode_minus"] = False
## 从文件导入数据
data = type_meanvisit
## 提取其中的columns数组，视为数据的标签
name = []
for style in type_meanvisit['type']:
    name.append(style[2:])
## 提取其中的values数组，数据的存在位置
values = type_meanvisit["avg(visit)"]
#print(name)
#print(values)
# 刻度标签
labels = name
## 设置画布
plt.figure(figsize=(20,10))
## 绘制直方图
plt.bar(range(len(type_meanvisit)), values, width=0.8, color="r")
## 添加横轴标签
plt.xlabel("类型")
## 添加y轴名称
plt.ylabel("平均访问量")
## 设置x轴刻度
plt.xticks(range(len(type_meanvisit)), labels)
## 添加图表标题
plt.title("笑话类型及其平均访问量直方图")
plt.xticks(range(len(labels)),name[:], rotation=45)
## 保存图形到文件
plt.savefig("笑话类型及其平均访问量直方图.png")
## 显示图形
plt.show()
```

## 笑话类型对应笑话平均访问量的饼图如下

```python
plt.figure(figsize=(20,20))
plt.pie(values, labels=name[:], autopct="%1.2f%%")
## 保存图形到文件
plt.savefig("笑话类型及其平均访问量饼图.png")
plt.show()
```
